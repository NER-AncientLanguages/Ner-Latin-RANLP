```python
df_juv = pd.read_csv('LASLA_comparison_GOLD_Juv_BERTdiff.csv', index_col=0)
df_cic = pd.read_csv('LASLA_comparison_GOLD_Cic_BERTdiff.csv', index_col=0)
df_tac = pd.read_csv('LASLA_comparison_GOLD_Tac_BERTdiff.csv', index_col=0)
```

```python
df_tac
```

```python
df_juv = df_juv[(df_juv['AGREED_ann_1'] == df_juv['ent_label_BERT'])&(df_juv['ent_label_BERT_hypopt'] == 'O')]
df_cic = df_cic[(df_cic['BIO_gold'] == df_cic['ent_label_BERT'])&(df_cic['ent_label_BERT_hypopt'] == 'O')]
df_tac = df_tac[(df_tac['BIO_gold'] == df_tac['ent_label_BERT'])&(df_tac['ent_label_BERT_hypopt'] == 'O')]

prose = df_cic['form'].values.tolist() + df_tac['form'].values.tolist()
poetry = df_juv['form'].values.tolist()
```

```python
pattern = r'\b|\b'.join(prose)
print(pattern)

to_check['idk'] = to_check.word.str.contains(pattern + '\b')
words_in_set = to_check[to_check['idk'] == True]
print(len(words_in_set['word'].unique()))
print(len(prose))
```

```python
pattern = r'\b|\b'.join(poetry)
print(pattern)

to_check['idk'] = to_check.word.str.contains(pattern + '\b')
words_in_set = to_check[to_check['idk'] == True]
print(len(words_in_set['word'].unique()))
print(len(poetry))
```
